```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Part a - Setup a single node with static transitions
## Four possible states: Susceptible, Exposed, Infected, or Recovered

```python
#Initial Values
S = 0.8
E = 0.09
I = 0.01
R = 0
```

```python
#Simulation Parameters
beta = 0.1
alpha = 0.01
ro = 0.05
etha = 0.02
```

```python
#Run simulation
total_time = 2000

s = np.zeros((total_time, 1))
e = np.zeros((total_time, 1))
i = np.zeros((total_time, 1))
r = np.zeros((total_time))

s[0] = S
e[0] = E
i[0] = I
r[0] = R

for t in range(total_time - 1):
    #print(t)
    s[t+1] = s[t] - beta*i[t]*s[t] + etha*r[t]
    e[t+1] = e[t] + beta*i[t]*s[t] - alpha*e[t]
    i[t+1] = i[t] + alpha*e[t] - ro*i[t]
    r[t+1] = r[t] + ro*i[t] - etha*r[t]
```

```python
#Plot results
plt.figure(1)

#fig = plt.plot(s, label='Susceptible')
plt.plot(e, label='Exposed')
plt.plot(i, label='Infected')
plt.plot(r, label='Recovered')
plt.xlabel('days')
plt.ylabel('Percentage of Population')
plt.legend()

plt.show()
```
